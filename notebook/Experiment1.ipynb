```python
import os, sys

%load_ext autoreload
%autoreload 2

PYTHON_PATHS = [ '../','/home/alena/repos/ros-bag-processing']
for path in PYTHON_PATHS:
    if path not in sys.path:
        sys.path.append(path)

import cv2
import numpy as np
import matplotlib.pyplot as plt

from extraction.orb import orb_feature_exctraction
from matching.fundamental_matrix import plot_good_matches, find_good_matches


from tqdm import tqdm
import onnxruntime as rt
rt.get_device()
```

```python
def calculate_good_matches(img1, img2, filtering='fm', matching='knn'):
    _, kp1, ds1 = orb_feature_exctraction(img1)
    _, kp2, ds2 = orb_feature_exctraction(img2)
    num, _, _ = find_good_matches(kp1, kp2, ds1, ds2 , i, filtering=filtering, matching=matching)
    return num
```

```python
def read_images(dataset, i):
    img1 = cv2.imread(dataset + "%06i.png"%i)
    img2 = cv2.imread(dataset + "%06i.png"%(i+1))
    return img1, img2
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
seqs = np.sort(os.listdir("/home/alena/Documents/dark-slam/datasets/original/images/hist"))

N = 10
filterings = ["fm", "fm", "em_calib", "em_calib", "homo"]
matchings = ["knn", "bf", "knn", "bf", "knn"]

for i in range(len(filterings)):
    filtering = filterings[i]
    matching = matchings[i]
    print("FILTERING: ", filtering)
    print("MATCHING: ", matching)
    for seq in seqs:

        dataset = f"/home/alena/Documents/dark-slam/datasets/original/images/raw/{seq}/"
        dataset_hist = f"/home/alena/Documents/dark-slam/datasets/original/images/hist/{seq}/"
        dataset_gan = f"/home/alena/Documents/dark-slam/datasets/original/images/gan/{seq}/"
        dataset_gamma2 = f"/home/alena/Documents/dark-slam/datasets/original/images/gamma2/{seq}/"
        dataset_gamma4 = f"/home/alena/Documents/dark-slam/datasets/original/images/gamma4/{seq}/"
        images_names = np.sort(os.listdir(dataset_gamma4))
        good_matches_original = []
        good_matches_gamma = []
        good_matches_gan = []
        good_matches_hist = []
        
        for i in tqdm(range(images_names.shape[0] - 1)):
            # print(i)
            img1, img2 = read_images(dataset, i)
            num = calculate_good_matches(img1, img2, filtering=filtering, matching=matching)
            good_matches_original.append(num)

            #GAN
            img_gan1, img_gan2 = read_images(dataset_gan, i)
            num = calculate_good_matches(img_gan1, img_gan2, filtering=filtering, matching=matching)
            good_matches_gan.append(num)

            #HIST
            img_hist1, img_hist2 = read_images(dataset_hist, i)
            num = calculate_good_matches(img_hist1, img_hist2, filtering=filtering, matching=matching)
            good_matches_hist.append(num)

            #GAMMA
            img_gamma1, img_gamma2 = read_images(dataset_gamma2, i)
            num = calculate_good_matches(img_gamma1, img_gamma2, filtering=filtering, matching=matching)
            good_matches_gamma.append(num)
        
        plt.subplots(figsize=(10, 5))
        plt.plot(running_mean(good_matches_hist, N))
        plt.plot(running_mean(good_matches_gan, N))
        plt.plot(running_mean(good_matches_gamma, N))
        plt.plot(running_mean(good_matches_original, N))
        plt.title(seq + " " + filtering + " " + matching)
        plt.legend(["hist","gan","gamma","orig"])
        plt.show();
        
        print("Orig: ", sum(good_matches_original) / len(good_matches_original))
        print("Hist: ", sum(good_matches_hist) / len(good_matches_hist))
        print("GAN: ", sum(good_matches_gan) / len(good_matches_gan))
        print("Gamma2: ", sum(good_matches_gamma) / len(good_matches_gamma))
```
